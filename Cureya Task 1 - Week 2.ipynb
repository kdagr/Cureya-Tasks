```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
happiness_data = pd.read_csv('World Happiness Report 2021/world-happiness-report-2021.csv')
```

```python
happiness_data.head()
```

```python
X = happiness_data['Country name']
Y = happiness_data['Logged GDP per capita']
Z = happiness_data['Ladder score']
plt.figure(figsize=(30,4))
plt.plot(X,Y)
plt.plot(X,Z)
plt.xticks(rotation = 90) 
plt.title('Country Name vs (Scores and GDP)')
plt.xlabel('Countries')
plt.legend(['Ladder Score','GDP'])
plt.show()
```

```python
#here we can check the relationship between ladderscore and GDP by directly plotting them against each other
X = happiness_data['Ladder score']
Y = happiness_data['Logged GDP per capita']
plt.figure(figsize=(8,4))
plt.plot(X,Y)
plt.title('score vs GDP')
plt.xlabel('Ladder Score')
plt.ylabel('GDP')
plt.show()
```

```python
# creating a figure and axis 
fig, ax = plt.subplots() 
X = happiness_data['Ladder score']
Y = happiness_data['Logged GDP per capita']
plt.axis([2, 8.5, 6, 12])
#creating bar graph
ax.bar(X, Y) 
# set title and labels 
ax.set_title('Score vs GDP') 
ax.set_xlabel('Ladder score') 
ax.set_ylabel('GDP')
```

CONCLUSION:
1) Countries with higher Ladder score usually have higher GDP
2) But countries with higher GDP does not neccessarily have higher ladder score
3) Countries with lower GDP generally have low ladder score

```python
df = happiness_data.rename(columns = {'Regional indicator':'Regions','Logged GDP per capita':'GDP',
                            'Ladder score':'score','Healthy life expectancy':'life_exp'}, inplace = False)
df.head()
```

```python
WE = df[df.Regions == 'Western Europe' ]
SA = df[df.Regions == 'South Asia']
SEA = df[df.Regions == 'Southeast Asia']
plt.figure(figsize=(8,4))

plt.plot(WE.life_exp,WE.GDP,'o')
plt.plot(SA.life_exp,SA.GDP,'o')
plt.plot(SEA.life_exp,SEA.GDP,'o')
plt.title('life Expectancy vs GDP')
plt.xlabel('Life Expentancy')
plt.ylabel('GDP')
plt.legend(['Western Europe','South Asia','South East Asia'])
plt.show()
```

Conclusion: 
1) As GDP increases, Life expentancy also usually increases
2) Southeast Asian and South Asian countries have lower life expentancy as compared to Western European countries
