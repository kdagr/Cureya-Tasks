```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('titanic/train.csv')
test = pd.read_csv('titanic/test.csv')
```

```python
train.head()
```

```python
train.info()
```

```python
train.describe()
```

```python
df_train = train
df_train
```

```python
to_drop = ['Name', 'SibSp','Parch', 'Ticket', 'Fare', 'Cabin' ,'Embarked']
train.drop(columns=to_drop, inplace = True)
train
```

```python
train.info()
```

```python
#as there are approx hundread null entires for feature Age, I will just drop them
train = train.dropna()
```

```python
train
```

```python
X = train[['Pclass','Sex','Age']]
Y = train.Survived
```

```python
X.replace(to_replace ="male", value = "1", inplace = True)
X.replace(to_replace ="female", value = "0", inplace = True)
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X_train,y_train)
```

```python
y_pred = logreg.predict(X_test)
```

# Model Evaluation using Confusion Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import metrics
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
